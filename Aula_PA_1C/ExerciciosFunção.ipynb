# **Exercico 2**

```python
##Conversor de Temperatura
def celsius_para_fahrenheit():
  celsius = float(input("Digite a temperatura em Celsius: "))
  fahrenheit = (celsius * 9/5) + 32
  return (f"{celsius} é igual a {fahrenheit} em fahrenheit")
```

# **Exercicio 6**

```python
##Cálculo de Desconto
def calcular_desconto(valor_original, percentual_desconto):
  valor_final = valor_original * (1 - percentual_desconto / 100)
  return valor_final
```

```python
calcular_desconto(100, 10)
```

# **Exercicio 7**

```python
##Maior e Menor
def maior_e_menor():
  lista = []
  for i in range(5):
    numero = int(input("Digite um número: "))
    lista.append(numero)
  maior = max(lista)
  menor = min(lista)
  return (maior, menor)
```

```python
maior_e_menor()
```

# **Exercicio 8**

```python
##Calculadora Simples
def calculadora_simples():
  operação = input("Digite a forma de calculo (+, -, *, /)")
  num1 = float(input("Digite o primeiro número: "))
  num2 = float(input("Digite o segundo número: "))
  if operação == "+":
    resultado1 = num1 + num2
    print(resultado1)
  elif operação == "-":
    resultado2 = num1 - num2
    print(resultado2)
  elif operação == "*":
    resultado3 = num1 * num2
    print(resultado3)
  elif operação == "/":
    resultado4 = num1 / num2
    print(resultado4)
  else:
    print("Operação inválida")
```

```python
calculadora_simples()
```

```python
##Mensagem Automática
def enviar_email(destinatario, assunto, corpo="Este é um email automático."):
  destinatario = input("Digite o destinatário: ")
  assunto = input("Digite o assunto: ")
  corpo = input("Digite o corpo: ")
  print(f"Enviando email para: {destinatario}")
  print(f"Assunto: {assunto}")
  print(f"Corpo: {corpo}")
  print("Email enviado com sucesso!")
```
